```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import data into notebook
diabetes = pd.read_csv('https://raw.githubusercontent.com/niteen11/DataAnalyticsAcademy/master/Python/dataset_diabetes/diabetic_data.csv')
```

```python
#Drop na values
diabetes.dropna(inplace=True)
```

```python
diabetes.shape
```

```python
#Create smaller data to view gender and number of medications correlation
data = diabetes.groupby(['gender'])['num_medications'].sum()
```

```python
data.head()
```

```python
#Create visualization showing correlation between gener and number of medications
diabetes.groupby(['gender'])['num_medications'].count().sort_values(ascending=False).plot(kind='barh')
```

```python
#Create smaller data to view race and max A1C correlation
A1cdata = diabetes.groupby(['race'])['A1Cresult'].max()
```

```python
A1cdata.head()
```

```python
#Create smaller data to view race and min A1C correlation
A1cdata = diabetes.groupby(['race'])['A1Cresult'].min()
```

```python
A1cdata.head()
```

```python
#Create visualization showing correlation between gener and number of medications
diabetes.groupby(['race'])['A1Cresult'].count().sort_values(ascending=False).plot(kind='bar')
```

```python
#Compare readmission rates among different groups
diabetes.groupby(['age', 'gender', 'race'])['readmitted'].value_counts(normalize=True).unstack()
```

```python
#Visualization of race comparison
plt.subplot(1, 3, 3)
sns.countplot(x='race', hue='readmitted', data=diabetes)
plt.title('Race Comparison')
plt.xlabel('Race')
plt.ylabel('Count')
plt.legend(title='Readmitted', loc='upper right')
```

```python
#Distribution of readmission
plt.figure(figsize=(8, 6))
sns.countplot(x='readmitted', data=diabetes)
plt.title('Distribution of Readmission')
plt.xlabel('Readmitted')
plt.ylabel('Count')
plt.show()
```

```python
#Perform demographics comparison
diabetes.groupby('readmitted')[['age', 'gender', 'race']].describe()
```

```python
#Analyze treatments received
diabetes.groupby('readmitted')['diabetesMed'].value_counts(normalize=True).unstack()
```

```python
# Age comparison
plt.subplot(1, 3, 1)
sns.countplot(x='age', hue='readmitted', data=diabetes)
plt.title('Age Comparison')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='Readmitted', loc='upper right')
```

```python
#Race comparison
plt.subplot(1, 3, 3)
sns.countplot(x='race', hue='readmitted', data=diabetes)
plt.title('Race Comparison')
plt.xlabel('Race')
plt.ylabel('Count')
plt.legend(title='Readmitted', loc='upper right')
plt.tight_layout()
plt.show()
```
